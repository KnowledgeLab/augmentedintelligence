We can first load the data, uncomment out the one you want. We are back to one `DataFrame` so things are simple.

```python
#df = ai.condense4.loadNewsGroups()
#df = ai.condense4.loadSenate()
df = ai.condense4.loadReddit()
```

Now we can make a word cooccurrence graph

```python
g = ai.condense4.wordCooccurrence(df['normalized_sents'].sum())
```

To visualize we need to reduce the size of the graph, this function will do that, play with the numbers and see how the end result is affected.

```python
g = ai.condense4.reduceGraph(g, maxEdge = 1000, maxNode = 200)
```

Now we can look at a few properies of the graph:

```python
ai.condense4.displayBetweeness(g)
#ai.condense4.displayEigen(g)
#ai.condense4.displayCloseness(g)
#ai.condense4.displayDegree(g)
```

Or the neighbours of a single node

```python
IPython.display.Image(ai.condense4.displayNeighbours(g, 'on'))
```
